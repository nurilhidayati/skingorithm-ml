```python
import os
import re
```

```python
#ngurutin rename yang ada id nya
def sorted_alphanumeric(data):
    convert = lambda text: int(text) if text.isdigit() else text.lower()
    alphanum_key = lambda key: [ convert(c) for c in re.split('([0-9]+)', key) ] 
    return sorted(data, key=alphanum_key)
```

```python
path = os.chdir('C:/Users/DPJSIG04-PRO/Documents/capstone/project/panda/image_tif/')
filename = os.listdir(path)
srtdfile = sorted_alphanumeric(filename)

id = 1

for file in srtdfile:
    str_id = str(id)
    zero_id = str_id.zfill(3)
    new = "{}_eye.PNG".format(zero_id)
    print("renaming: \n {}".format(file))
    print("\n into: \n {}".format(new))
    os.rename(file, new)
    id += 1
```
